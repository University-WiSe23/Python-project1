Aufgabe 4 (Typ C)
Sie haben einen Kapitalbetrag von 66,66 EURO, den sie mit einem Zinssatz von 8 % für 10 Jahre anlegen. 
Schreiben sie ein Programm, welches den Endwert nach diesen 10 Jahren errechnet und in einem Antwortsatz am Bildschirm ausgibt. 
Dabei soll der Endwert auf 2 Stellen hinter dem Komma gerundet sein.

    Tipp 1: Der Endwert ergibt sich nach der Formel: endWert = kapitalBetrag * (1+zinsSatz/100)**anzahlJahre
    Tipp 2: Der Funktionsaufruf round(x,2) hat den auf zwei Stellen hinter dem Komma gerundeten Wertvon x als Rückgabewert.

a) Berechnen Sie endWert zunächst ohne Aufruf einer Funktion.

b) Verwenden Sie in einer Alternativlösung zur endWert-Berechnung bei der Potenzberechnung (1+zinsSatz/100) anzahlJahre die 
Potenzmethode pow aus dem Paket math.
[pow erwartet beim Aufruf zwei Argumente: pow(basis, exponent); in dieser Aufgabe wäre also (1+zinsSatz/100) die Basis und anzahlJahre der Exponent.]

c) Ergänzen Sie Ihre Lösung von b), indem Sie ein einfaches Liniendiagramm erstellen: 
Es sollen dabei nur der Anfangsbetrag im Jahr 0 und der Endbetrag im Jahr 10 als Wertevorgegeben werden. 
Die x-Achse soll mit „Jahre“, die y-Achse mit „Betrag“ beschriftetwerden; die Gesamtgraphik soll die Überschrift „Kapitalbetrag nach Jahren“ erhalten.
(Verwenden Sie dazu matplotlib.)

```python
#a)
anzahlJahre=10
kapitalBetrag=66.66
zinsSatz=8/100

endWert=kapitalBetrag*((1+zinsSatz)**anzahlJahre)
endWert_gerundet=round(endWert,2)
print("Der Endwert beträgt: "+str(endWert_gerundet))
```

```python
#b)
anzahlJahre=10
kapitalBetrag=66.66
zinsSatz=8/100
import math
endWert=kapitalBetrag*(pow((1+zinsSatz),anzahlJahre))
endWert_gerundet=round(endWert,2)
print("Der Endwert beträgt: "+str(endWert_gerundet))
```

```python
#c)
anzahlJahre=10
kapitalBetrag=66.66
zinsSatz=8/100
import math
endWert=kapitalBetrag*(pow((1+zinsSatz),anzahlJahre))
endWert_gerundet=round(endWert,2)
print("Der Endwert beträgt: "+str(endWert_gerundet))

import matplotlib as plt
Anfangsbetrag=[66.66, 143.91]
import matplotlib.pyplot as plt

Anfangsbetrag = [66.66, endWert_gerundet]
Jahre = [0, 10]

plt.plot(Jahre, Anfangsbetrag)
plt.xlabel('Jahre')
plt.ylabel('Betrag (in mio.)')
plt.title('Kapitalbetrag nach Jahren')

plt.show()
```
